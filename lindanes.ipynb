```python
from rdkit import Chem
from rdkit.Chem import PandasTools
import mols2grid
from sigma import SigmaCalculator
```

Read the lindanes as 3D structures from and SD file

```python
df = PandasTools.LoadSDF("lindanes.sdf")
```

View the 3D structures

```python
mols2grid.display(df,mol_col="ROMol",n_cols=4,selection=False)
```

Assign stereochemistry from the 3D structures

```python
df.ROMol.apply(Chem.AssignStereochemistryFrom3D)
df['SMILES'] = df.ROMol.apply(Chem.MolToSmiles)
```

Display the 2D structures

```python
cols = [x for x in df.columns if x != 'ROMol']
mols2grid.display(df[cols],subset=["img","ID"],n_cols=4,selection=False)
```

```python
sigma_calculator = SigmaCalculator()
```

Define a 2D molecule for display

```python
df['mol'] = df.SMILES.apply(Chem.MolFromSmiles)
```

Calculate sigma

```python
df['sigma'] = df.mol.apply(sigma_calculator.calc_sigma_new)
```

Display molecules with the name and sigma values below each structure

```python
cols = [x for x in df.columns if x != 'ROMol']
mols2grid.display(df[cols],subset=["img","ID","sigma"],n_cols=4,selection=False)
```
