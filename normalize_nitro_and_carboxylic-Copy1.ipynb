```python
from rdkit import Chem
from sigma import SigmaCalculator
from rdkit.Chem import rdChemReactions
```

### Nitro groups

Define a function to normalize ntiro groups

```python
def normalize_nitro(mol):
    nitro_norm = rdChemReactions.ReactionFromSmarts("[N+:1](=[O:2])[O-:3]>>[N:1](:[O:2])(:[O+0:3])")
    new_mol = Chem.Mol(mol)
    while nitro_norm.RunReactantInPlace(new_mol):
        pass
    Chem.SanitizeMol(new_mol)
    return new_mol
```

Create paradinitrobenzene

```python
para_dinitrobenzene = Chem.MolFromSmiles("[O-][N+](=O)c1ccc(cc1)[N+]([O-])=O")
```

```python
para_dinitrobenzene
```

Calculate sigma using the original method

```python
calc = SigmaCalculator()
calc.calc_sigma_original(para_dinitrobenzene)
```

Calculate sigma using the new 3D method

```python
calc.calc_sigma_new(para_dinitrobenzene)
```

Normalize the nitro groups

```python
para_dinitrobenzene_norm = normalize_nitro(para_dinitrobenzene)
para_dinitrobenzene_norm
```

Calculate again

```python
calc.calc_sigma_new(para_dinitrobenzene_norm)
```

Repeat the calculation with the original method

```python
calc.calc_sigma_original(para_dinitrobenzene_norm)
```

### Carboxylic acids

Define a function to normalize carboxylic acids

```python
def normalize_carboxylic_acid(mol):
    carboxylic_norm = rdChemReactions.ReactionFromSmarts("[C:1](=[O:2])[O:3]>>[C:1](:[O:2]):[O:3]")
    new_mol = Chem.Mol(mol)
    while carboxylic_norm.RunReactantInPlace(new_mol):
        pass
    Chem.SanitizeMol(new_mol)
    return new_mol
```

```python
terephthalic_acid = Chem.MolFromSmiles("OC(=O)c1ccc(cc1)C(O)=O")
```
