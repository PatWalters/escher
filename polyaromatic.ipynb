```python
import pandas as pd
from io import StringIO
from rdkit import Chem
from rdkit.Chem import PandasTools
import mols2grid
from sigma import SigmaCalculator
```

Input structures and names

```python
buff = """c1ccc2c3ccccc3Cc2c1 fluorene
c1ccc2c(c1)c3ccccc3[nH]2 carbazole
o2c1ccccc1c3c2cccc3 dibenzofuran
c1ccc2c(c1)c3ccccc3s2 dibenzothiophene
c1c(cccc1)Cc2ccccc2 diphenylmethane
c1ccc(cc1)Nc2ccccc2 diphenylamine
O(c1ccccc1)c2ccccc2 diphenyl_ether
S(c1ccccc1)c2ccccc2 diphenyl_sulphide"""
```

Read the data above into a dataframe

```python
ifs = StringIO(buff)
df = pd.read_csv(ifs,sep=" ",names=["SMILES","Name"])
df['mol'] = df.SMILES.apply(Chem.MolFromSmiles)
```

View the structures

```python
mols2grid.display(df,mol_col='mol',subset=["img","Name"],n_cols=4,selection=False)
```

Instantiate a sigma calculator and calculate sigma

```python
sigma_calculator = SigmaCalculator()
df['sigma'] = df.mol.apply(sigma_calculator.calc_sigma_new)
```

Display the results with structure, name, and sigma

```python
mols2grid.display(df,mol_col='mol',subset=["img","Name","sigma"],n_cols=4,selection=False)
```
